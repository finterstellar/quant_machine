```
import pandas as pd
import pandas_datareader.data as web
import requests
from io import StringIO
```

```
%%time
df = pd.DataFrame()
df = web.DataReader(['DGS10', 'T10Y3M', 'ICSA', 'WALCL', 'INDPRO', 'T10YIE'], 'fred')
df.fillna(method='ffill', inplace=True)
print(df)
```

```
%%time
df = pd.DataFrame()
symbols = ['DGS10', 'T10Y3M', 'ICSA', 'WALCL', 'INDPRO', 'T10YIE']

def crawl_fred(s):
    url = 'https://fred.stlouisfed.org/graph/fredgraph.csv?id={}'.format(s)
    r = requests.get(url)
    rst = pd.read_csv(StringIO(r.text), index_col=0, parse_dates=True, header=None,skiprows=1, names=['DATE', s], na_values=".")
    return rst
df = pd.concat([crawl_fred(s) for s in symbols], axis=1, join="outer")
df.fillna(method='ffill', inplace=True)
print(df)
```

```
%%time
df = pd.DataFrame()
symbols = ['DGS10', 'T10Y3M', 'ICSA', 'WALCL', 'INDPRO', 'T10YIE']

for s in symbols:
    url = 'https://fred.stlouisfed.org/graph/fredgraph.csv?id={}'.format(s)
    r = requests.get(url)
    rst = pd.read_csv(StringIO(r.text), index_col=0, parse_dates=True, header=None,skiprows=1, names=['DATE', s], na_values=".")
    df = pd.concat([df,rst], axis=1)
df.fillna(method='ffill', inplace=True)
print(df)
```
