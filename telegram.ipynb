```
import telegram
```

```
msg = 'Telegram message'
token = '1130429321:AAFzLWZRnMLy5T6qT7yMilLdthUJvfLxXV0'
```

```
def telegram_msg(token, msg):
    bot = telegram.Bot(token=token)
    # 대화방 멤버들의 id 수집
    id = set()
    for r in bot.getUpdates():
        id.add(r.message.chat['id'])
    # 대화방 멤버들에게 메시지 전송
    for i in id:
        bot.sendMessage(chat_id=i, text=msg)
```

```
telegram_msg(token, msg)
```
